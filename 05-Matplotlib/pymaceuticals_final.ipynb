```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mdd_tl = pd.read_csv(mouse_drug_data_to_load)
ctd_tl = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mdd_ctd_tl = pd.merge(mdd_tl,ctd_tl, on="Mouse ID")

# Display the data table for preview
mdd_ctd_tl.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
group_mdd_ctd_tl = mdd_ctd_tl.groupby(['Drug','Timepoint']).mean().reset_index()

# Convert to DataFrame
mdd_ctd_df = group_mdd_ctd_tl.drop(columns='Metastatic Sites')

# Preview DataFrame
mdd_ctd_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stats_mdd_ctd = mdd_ctd_tl.groupby(['Drug','Timepoint']).sem().reset_index()

# Convert to DataFrame
stats_mdd_ctd_df = stats_mdd_ctd.drop(columns=['Metastatic Sites','Mouse ID'])

# Preview DataFrame
stats_mdd_ctd_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
bydrug_mean_df = pd.pivot_table(mdd_ctd_df,index=['Timepoint'],columns='Drug',values=f'Tumor Volume (mm3)')
bydrug_sem_df = pd.pivot_table(stats_mdd_ctd_df,index=['Timepoint'],columns='Drug',values=f'Tumor Volume (mm3)')

# Preview that Reformatting worked
bydrug_mean_df.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis=np.arange(0,50,5)
plt.figure(figsize=(10,10))

#Picked Highest, Middle Highest, Middle, Middlow Low, then Lowest plots
capomulin = plt.errorbar(x_axis, bydrug_mean_df['Capomulin'], yerr=bydrug_sem_df['Capomulin'],markersize=5, marker='o', color='black',linewidth=1.5, label='Capomulin')
ketapril = plt.errorbar(x_axis, bydrug_mean_df['Ketapril'], yerr=bydrug_sem_df['Ketapril'],markersize=5, marker='o', color='red',linewidth=1.5, label='Ketapril')
placebo = plt.errorbar(x_axis, bydrug_mean_df['Placebo'], yerr=bydrug_sem_df['Placebo'],markersize=5, marker='o', color='blue',linewidth=1.5, label='Placebo')
ramicane = plt.errorbar(x_axis, bydrug_mean_df['Ramicane'], yerr=bydrug_sem_df['Ramicane'],markersize=5, marker='o', color='orange',linewidth=1.5, label='Ramicane')
ceftamin = plt.errorbar(x_axis, bydrug_mean_df['Ceftamin'], yerr=bydrug_sem_df['Ceftamin'],markersize=5, marker='o', color='green',linewidth=1.5, label='Ceftamin')

plt.xlabel(f'Time (Days)',fontsize=15)
plt.ylabel(f'Tumor Volume (mm3',fontsize=15)
plt.title('Tumor Response to Treatment',fontsize=15)
plt.grid(linestyle='-', linewidth=.5)
plt.legend(loc=2,prop={'size': 14})

# Save the Figure
plt.savefig('Tumor_Response_To_Treatment.png')

#Show the Figure
plt.show()
print(f'Observation #1: Ramicane and Capomulin responded in reducing tumor sizes.')
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_mdd_ctd_tl = mdd_ctd_tl.groupby(['Drug','Timepoint']).mean()

# Convert to DataFrame
met_mdd_ctd_tl = met_mdd_ctd_tl.drop(columns=f'Tumor Volume (mm3)')

# Preview DataFrame
met_mdd_ctd_tl.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_mdd_ctd_sem = mdd_ctd_tl.groupby(['Drug','Timepoint']).sem()
# Convert to DataFrame
met_mdd_ctd_sem2 = met_mdd_ctd_sem.drop(columns=[f'Tumor Volume (mm3)','Mouse ID'])
# Preview DataFrame
met_mdd_ctd_sem2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

met_bydrug_mean_df = pd.pivot_table(met_mdd_ctd_tl,index=['Timepoint'],columns='Drug',values=f'Metastatic Sites')
met_bydrug_sem_df = pd.pivot_table(met_mdd_ctd_sem2,index=['Timepoint'],columns='Drug',values=f'Metastatic Sites')

# Preview that Reformatting worked
met_bydrug_mean_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(10,10))

capomulin2 = plt.errorbar(x_axis, met_bydrug_mean_df['Capomulin'], yerr=met_bydrug_sem_df['Capomulin'],marker='o', markersize=5, color='black',linewidth=1.5, label='Capomulin')
infubinol2 = plt.errorbar(x_axis, met_bydrug_mean_df['Infubinol'], yerr=met_bydrug_sem_df['Infubinol'],marker='o', markersize=5, color='red',linewidth=1.5, label='Infubinol')
ketapril2 = plt.errorbar(x_axis, met_bydrug_mean_df['Ketapril'], yerr=met_bydrug_sem_df['Ketapril'],marker='o', markersize=5, color='green', linewidth=1.5, label='Ketapril')
placebo2 = plt.errorbar(x_axis, met_bydrug_mean_df['Placebo'], yerr=met_bydrug_sem_df['Placebo'],marker='o', markersize=5, color='blue', linewidth=1.5, label='Placebo')
ramicane2 = plt.errorbar(x_axis, met_bydrug_mean_df['Ramicane'], yerr=met_bydrug_sem_df['Ramicane'], marker='o', markersize=5, color='orange', linewidth=1.5, label='Ramicane')

plt.xlabel(f'Treatment Duration (Days)',fontsize='15')
plt.ylabel(f'Metastatic Sites',fontsize='15')
plt.title('Metastatic Spread During Treatment',fontsize='15')
plt.grid(linestyle='-', linewidth=.5)
plt.legend(loc=2, prop={'size': 14})

# Save the Figure
plt.savefig('Metastatic_Spred_During_Treatment.png')

# Show the Figure
plt.show()
print(f'Observation #2: Ramicane and Capomulin had the lowest Metastatic Spread of all treatments.')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_group = mdd_ctd_tl.groupby(['Drug','Timepoint'])
survival_count = survival_group['Mouse ID'].count()
# Convert to DataFrame
survival_df = pd.DataFrame(survival_count).reset_index()
survival_df = survival_df.rename(columns={'Mouse ID':'Mouse Count'})
# Preview DataFrame
survival_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_bydrug_df = pd.pivot_table(survival_df,index=['Timepoint'],columns='Drug',values=f'Mouse Count')
# Preview the Data Frame
survival_bydrug_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.figure(figsize=(8,8))

new_df = {}
new_df['Timepoint'] = survival_bydrug_df.index

col_1 = (survival_bydrug_df['Capomulin']/25)*100
col_2 = (survival_bydrug_df['Ceftamin']/25)*100
col_3 = (survival_bydrug_df['Infubinol']/25)*100
col_4 = (survival_bydrug_df['Ketapril']/25)*100
col_5 = (survival_bydrug_df['Naftisol']/25)*100
col_6 = (survival_bydrug_df['Placebo']/25)*100
col_7 = (survival_bydrug_df['Propriva']/26)*100
col_8 = (survival_bydrug_df['Ramicane']/25)*100
col_9 = (survival_bydrug_df['Stelasyn']/26)*100
col_10 = (survival_bydrug_df['Zoniferol']/25)*100
new_df['Capomulin'] = col_1
new_df['Ceftamin'] = col_2
new_df['Infubinol'] = col_3
new_df['Ketapril'] = col_4
new_df['Naftisol'] = col_5
new_df['Placebo'] = col_6
new_df['Propriva'] = col_7
new_df['Ramicane'] = col_8
new_df['Stelasyn'] = col_9
new_df['Zoniferol'] = col_10


final_df = pd.DataFrame(new_df).astype(int)
final_df = final_df.drop(columns='Timepoint')

capomulin2 = plt.errorbar(x_axis, final_df['Capomulin'],marker='o', markersize=5, color='black',linewidth=1.5, label='Capomulin')
infubinol2 = plt.errorbar(x_axis, final_df['Infubinol'],marker='o', markersize=5, color='red',linewidth=1.5, label='Infubinol')
ketapril2 = plt.errorbar(x_axis, final_df['Ketapril'],marker='o', markersize=5, color='green',linewidth=1.5, label='Ketapril')
placebo2 = plt.errorbar(x_axis, final_df['Placebo'],marker='o', markersize=5, color='blue',linewidth=1.5, label='Placebo')
ramicane2 = plt.errorbar(x_axis, final_df['Ramicane'],marker='o', markersize=5, color='orange',linewidth=1.5, label='Ramicane')

plt.xlabel(f'Time (Days)',fontsize='15')
plt.ylabel(f'Survival Rate (%)',fontsize='15')
plt.title('Survival During Treatment',fontsize='15')
plt.grid(linestyle='-', linewidth=.5)
plt.legend(loc=3, prop={'size': 14})

# Save the Figure
plt.savefig('Survival_During_Treatment.png')

# Show the Figure
plt.show()
print(f'Observation #3: Ramicane and Capomulin had the highest survival rates during treatment.')
```
